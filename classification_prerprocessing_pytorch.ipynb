#data preprocessing

```python
cat_miss_cols= [ 	'experience', 'last_new_job' ,'gender',	'enrolled_university',	'education_level',	'major_discipline',	'company_size' , 'company_type']
```

```python
for col in cat_miss_cols:
  print(df[col].unique)
```

```python
df[cat_miss_cols] = df[cat_miss_cols].fillna('NA')
```

```python
columns_with_nans = df.columns[df.isna().any()].tolist()
print(columns_with_nans)
```

```python
for colmns in columns_with_nans:
  df[col] = pd.to_numeric(df[col].replace('nan', np.Nan))
```

```python
df.info()
```

```python
columns_with_nans = df.columns[df.isna().any()].tolist()
print(columns_with_nans)
```

```python
cat_cols = df.select_dtypes(include='O').columns
cat_cols
```

```python
for col in cat_cols:
  print(col)
  print(df[col].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oc = OrdinalEncoder(
    categories = [['Primary School', 'High School', 'NA', 'Graduate', 'Masters' , 'Phd']])
df[['education_level']] = oc.fit_transform(df[['education_level']])
oc.categories_
```

```python
oc2 = OrdinalEncoder(
    categories = [['<10', '10/49', '50-99', '100-500', '500-999', 'NA', '1000-4999', '5000-9999', '10000+']])
df[['company_size']] = oc2.fit_transform(df[['company_size']])
oc2.categories_
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```
